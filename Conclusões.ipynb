# 1. Diagnóstico de Performance

A análise dos dados (set/2021 a mar/2022) revelou uma clara tendência de recuperação no engajamento a partir de novembro de 2021. Embora o volume inicial de dados fosse limitado para análises sazonais precisas, a segmentação por tipo de conteúdo permitiu identificar padrões consistentes de sucesso. O formato Foto consolidou-se como o pilar de estabilidade da conta, enquanto o Reels apresenta o maior potencial de escala, apesar da amostragem reduzida.

# 2. O "Post de Ouro" (Diretrizes Práticas)

Identificamos que o engajamento não é aleatório. Existe uma combinação de atributos que maximiza a probabilidade de sucesso. Para otimizar os resultados, as publicações devem priorizar:

Aparência: `Fotos estáticas` com a presença de `pessoas` (rostos reais aumentam drasticamente a conexão).

Contexto: Conteúdos atrelados a `Campanhas` e com a tag `Trends` ou `Promoções` e `Influenciadores`.

Formato: **Evitar o uso de Carrossel**, que apresentou performance inferior às fotos únicas no período analisado.

# 3. Próximos Passos e Experimentos

Para o próximo trimestre, recomenda-se:

Escalar o Reels e Vídeos: Realizar testes controlados com as tags de `Trends` e `Datas Comemorativas` para validar se o alto engajamento inicial se mantém em maior volume juntamente a presença de `Pessoas` e `Campanhas`.

Estudar melhor o motivo de postagens `Sem Tags` também apresentarem bom desempenho. Apresentam algum conteúdo único? Alguma marca específica dessa loja?

Ajuste de Mix: Substituir postagens de `Produtos` (estáticas/sem pessoas) por fotos ambientadas com pessoas utilizando os produtos.

Monitoramento: Manter a coleta de dados para reavaliar a eficácia das `Fotos de Promoção`, garantindo que o excesso desse tipo de conteúdo não sature a audiência
