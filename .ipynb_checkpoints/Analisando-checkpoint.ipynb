```python
colunas_cat_tags = ['Tags', 'Carrossel', 'Pessoas', 'Campanhas']
for coluna in colunas_cat_tags:
    df_tags[coluna] = df_tags[coluna].astype('category')
```

# Análise Temporal

## Gráfico

```python
df.groupby(df['Data'].dt.to_period('M'))[['Interacoes', 'Curtidas', 'Comentários']].mean().plot(kind='line', marker='s', figsize=(12,6))
plt.title('Evolução Mensal: Interações, Curtidas e Comentários')
plt.xlabel('Mês/Ano')
plt.ylabel('Média de Engajamento')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(['Total Interações', 'Curtidas', 'Comentários'])
plt.show()
```

## Conclusão da Análise Temporal

A falta de dados impede inferir muitas informações sobre comportamento sazonal, tendo apenas a visão do início da presença online tendo um pico no início, uma queda e posteriormente outro pico ainda maior que o anterior, contudo a previsão apresenta um vale que pode ser corrigido com uma modificação na presença do perfil. <br>
Devido a pequena amostragem em relação ao período de queda e recuperação e a duração do período de recuperação, a fragmentação dos dados os tornam muito imprecisos para qualquer análise com precisão ser feita

# Conclusão Final

Devido a pouca quantidade de dados uma conclusão precisa se torna difícil, contudo, baseado nas informações cheguei nas seguintes conclusões

> `Vídeos`, `Reels`, `IGTV` possuem poucas amostras. `Reels` possuiu o maior desempenho geral. Válido futuros experimentos para maior acurácia dos dados. `Vídeos` com `Pessoas` e `Campanhas` apresentaram bom desempenho, assim como `Reels`, experimentos futuros são desejáveis. <br>
> `Fotos` com a maior amostra e a segunda opção em todas as categorias a torna a melhor opção no momento

> `Promoção`, `Influenciadores`, `Novos Produtos` são as opções de Tags mais seguras para seguir <br>
> `Datas Comemorativas`, `Trends`, `Sem Tags` são válidos experimentos

> A análise por meio da passagem de tempo não possui dados para inferir qualquer análise precisa e/ou informativa o suficiente, contudo a curvatura indica uma tendência de crescimento no engajamento do perfil nos últimos meses

<br>

`Fotos` é a opção mais segura para "`Tipo`", `Promoção` se mostrou a mais promissora para "`Tags`" porém não é viável durante muito tempo para as vendas tornando `Trends` a melhor opção rentável de maior rotatividade enquanto `Datas Comemorativa` para eventos pontuais <br>
`Reels`, `Datas Comemorativa` & `Trends` são os de melhor desempenho com menor amostra, testes necessários <br>
As publicações com `Pessoas` obtiveram maior desempenho <br>
As publicações com `Carrossel` não tiveram desempenho superior <br>
As publicações com `Campanhas` tem maior engajamento

Com as informações acima é possível inferir que `Fotos` com a tag `Trends` com `Pessoas` **presentes** com uma `Camapanha` atrelada a publicação

```python
categorias = ['Curtidas', 'Comentários', 'Interacoes']
media_geral = [df_carrossel[col].mean() for col in categorias]
filtro = (df_carrossel['Pessoas'] == 'S') & (df_carrossel['Campanhas'] == 'S')
media = [df_carrossel.loc[filtro, col].mean() for col in categorias]
x = np.arange(len(categorias))
largura = 0.2
fig, ax = plt.subplots(figsize=(4, 2.6))
rects1 = ax.bar(x - largura/2, media_geral, largura, label='Média Geral (Fotos)', color='#BDC3C7')
rects2 = ax.bar(x + largura/2, media, largura, label='Estratégia', color='#27AE60')
ax.set_ylabel('Quantidade Média', fontsize=7)
ax.set_title('Comparativo Detalhado: Performance Geral vs. Estratégia', fontsize=8)
ax.set_xticks(x)
ax.set_xticklabels(['Curtidas', 'Comentários', 'Total Interações'], fontsize=7)
ax.legend(loc='upper center', fontsize=6)
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.0f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, -2), 
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold',
                   fontsize=7)
autolabel(rects1)
autolabel(rects2)
plt.tight_layout()
plt.savefig('grafico_final.png', dpi=300)
plt.show()
```
